```python
from ultralytics import YOLO
from pathlib import Path
```

```python
!unzip -o SignBridge.v5-v4_hand.yolov8.zip
```

```python
model = YOLO('yolov8n-pose.pt')
```

```python
model.train(
    data=(Path.cwd() / 'data.yaml').absolute(),
    epochs=50,
    imgsz=640,
    batch=64,
    device=0
)
```

```python
model.save("sign-bridge.pt")
```
